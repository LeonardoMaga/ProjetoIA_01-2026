# Tratamento dos Dados


# Significados das Variáveis

ENERGY - Energy Consumption (kWh) **TARGET**

*   A quantidade total de energia elétrica consumida no período, medida em
quilowatts-hora.

HDD18_3 - Heating Degree Days below 18.3°C (°C)

*   Mede quantos graus a temperatura média externa ficou ABAIXO da temperatura de conforto (18,3°C), indicando que as pessoas ligaram aquecedores.

CDD0 - Cooling Degree Days above 0°C (°C)

*   Mede o quanto a temperatura externa ultrapassou a base de 0°C.

CDD10 - Cooling Degree Days above 10°C (°C)

*   Similar ao anterior, mas mede o quanto a temperatura externa ultrapassa a base de 10°C.

PRECTOT - Precipitation (mm/hour)

*   Precipitação  

RH2M - Relative Humidity at 2m (%)

*   A quantidade de vapor de água no ar em relação ao máximo que o ar conseguiria reter, medida a 2 metros de altura do solo (o padrão de estações meteorológicas).

T2M - Temperature at 2m (C)

*   A temperatura média do ar registrada a 2 metros da superfície terrestre.

T2M_MIN - Minimum Temperature at 2 m (C)

*   A temperatura mais baixa do ar registrada durante o período, a 2 metros do solo.

T2M_MAX - Maximum Temperature at 2 m (C)

*   A temperatura mais alta do ar registrada durante o período, a 2 metros do solo.

ALLSKY - All-Sky Surface Longwave Downward Irradiance (W/m^2)

*   Irradiância de Ondas Longas Descendente na Superfície em Céu Total (W/m²)
A quantidade de radiação térmica (calor) que a atmosfera e as nuvens emitem em direção à superfície da Terra, medida em watts por metro quadrado. A expressão "céu total" significa que leva em conta a presença real de nuvens no momento da medição.

HOLIDAY - Holidays in Spain

*   Feriados e fim de semana

# Carregamento dos Dados

```python
# Carrega os dados do csv em um data frame e indica que a coluna DATE possui valores que são datas
#df_raw = pd.read_csv('db_building_A.csv', parse_dates=['DATE'])

# Descomentar para utilizar os dados do prédio B
df_raw = pd.read_csv('db_building_B.csv', parse_dates=['DATE'])

print(f'Registros carregados : {len(df_raw):}')
print(f'Colunas              : {list(df_raw.columns)}')
print(f'Período              : {df_raw["DATE"].min()} --> {df_raw["DATE"].max()}')
df_raw.head()
```

```python
# Visão geral das estatísticas básicas
df_raw.describe().round(2)
```

# Identificando Problema nos Dados

```python
# Procurando valores ausentes
print('=== VALORES AUSENTES POR COLUNA ===')
ausentes = df_raw.isnull().sum() # Conta ausentes
ausentes_pct = (ausentes / len(df_raw) * 100).round(3) # Calcula porcentagem
resumo = pd.DataFrame({'Qtd. ausente': ausentes, '% do total': ausentes_pct})
print(resumo[resumo['Qtd. ausente'] > 0])
```

Possível solução: fazer uma interpolação linear para "consertar" já que não passa de 0.5% dos dados. Fonte: https://www.mdpi.com/2076-3417/11/17/7886


```python
# Procurando timestamps duplicados
duplicados = df_raw[df_raw.duplicated(subset='DATE', keep=False)].sort_values('DATE')
print(f'Timestamps duplicados encontrados: {df_raw["DATE"].duplicated().sum()}')
print()
print(duplicados[['DATE','ENERGY']].to_string(index=False))
```

```python
# Procura Gaps nos horários
df_sorted = df_raw.sort_values('DATE').drop_duplicates(subset='DATE') # Ordena o dataframe por timestamp
diffs = df_sorted['DATE'].diff()
gaps = diffs[diffs > pd.Timedelta('1h')] # Diferenças maiores que uma hora são gaps

print(f'Gaps temporais (diferença > 1h): {len(gaps)}')
print()
for idx, gap in gaps.items():
    print(f'  {df_sorted.loc[idx, "DATE"]}  -->  gap de {gap}')
```

```python
# Identifica possíveis erros de medição (valores menores que o consumo base)
suspeitos = df_raw[df_raw['ENERGY'] < 50] # 50 é um valor estipulado para o primeiro .csv, temos que decidir o valor pro segundo já que é um prédio com menos gasto
print(f'Registros com ENERGY < 50 kWh: {len(suspeitos)}')
print()
print(suspeitos[['DATE', 'ENERGY']].to_string(index=False))
```

# Aplicando Tratamento nos Dados

```python
# Percebemos que os registros duplicados seguim um padrão estranho, então durante uma pesquisa descobrimos que esse padrão ocorre pelo retorno do horário de verão europeu.
# Como os 2 valores são validos, tiramos uma média e substituimos por apenas 1 valor

df = df_raw.groupby('DATE').mean().reset_index()
df = df.sort_values('DATE').set_index('DATE')

print(f'Registros antes : {len(df_raw):,}')
print(f'Registros depois: {len(df):,}')
print(f'Duplicatas removidas: {len(df_raw) - len(df)}')
```

```python
# Restrutura os dados hora a hora, preenchendo os gaps de hora com Nan.

idx_completo = pd.date_range(start='2016-01-01 00:00',
                             end='2020-12-31 23:00',
                             freq='h')
df = df.reindex(idx_completo)
df.index.name = 'DATE'
```

```python
# Realiza a interpolação nos valores que estavam faltando

# Interpolação temporal para variáveis contínuas
colunas_continuas = ['ENERGY', 'ALLSKY', 'T2M', 'T2M_MIN', 'T2M_MAX', 'RH2M', 'PRECTOT']

for col in colunas_continuas:
    antes = df[col].isnull().sum()
    df[col] = df[col].interpolate(method='time')
    depois = df[col].isnull().sum()
    print(f'{col:10s}: {antes:3d} ausentes → {depois} após interpolação')

# Forward fill para variáveis diárias (mesmo valor para todas as horas do dia)
colunas_diarias = ['HDD18_3', 'CDD0', 'CDD10', 'HOLIDAY']

for col in colunas_diarias:
    antes = df[col].isnull().sum()
    df[col] = df[col].ffill()
    depois = df[col].isnull().sum()
    print(f'{col:10s}: {antes:3d} ausentes → {depois} após ffill')
```

```python
# Preenche os valores de ENERGY que são considerados como erros de medição com
# Nan e depois realiza a interpolação.

erros = df['ENERGY'] < 50
print(f'Registros com ENERGY < 50 kWh: {erros.sum()}')

df.loc[erros, 'ENERGY'] = np.nan
df['ENERGY'] = df['ENERGY'].interpolate(method='time').bfill() #bfill trata o erro do Ausentes em Energy, linha 1 é preenchida com o próximo valor

print(f'Ausentes restantes em ENERGY  : {df["ENERGY"].isnull().sum()}')
```

```python
# Verificação pós correções
print(f'Total de registros   : {len(df):,}')
print(f'Período              : {df.index.min()} → {df.index.max()}')
print(f'Total de ausentes    : {df.isnull().sum().sum()}')
print(f'ENERGY — mín: {df["ENERGY"].min():.2f}  máx: {df["ENERGY"].max():.2f}  média: {df["ENERGY"].mean():.2f} kWh')
print()
print('Dataset limpo.')
```

# Extração dos Dados Limpos

```python
input_path = 'db_building_A.csv'
# input_path = 'db_building_B.csv' #descomentar se utilizar o prédio B

output_path = input_path.replace('.csv', '_clean.csv')

df.reset_index().to_csv(output_path, index=False)

print(f'Base limpa exportada para: {output_path}')
```

# Criação das Features do DATA

```python
df['hour']        = df.index.hour
df['day_of_week'] = df.index.dayofweek   # 0 = segunda, 6 = domingo
df['month']       = df.index.month
df['is_weekend']  = (df.index.dayofweek >= 5).astype(int)

print("Features temporais criadas.")
print(df[['hour', 'day_of_week', 'month', 'is_weekend']].head(5))
```

```python
df['hour_sin']        = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos']        = np.cos(2 * np.pi * df['hour'] / 24)

df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

df['month_sin']       = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos']       = np.cos(2 * np.pi * df['month'] / 12)
```

```python
print(f'Colunas            : {list(df.columns)}')
print(f'Ausentes           : {df.isnull().sum().sum()}')

output_path = 'db_building_A_features.csv'
# output_path = 'db_building_B_features.csv'

df.reset_index().to_csv(output_path, index=False)
print(f'\nBase exportada para: {output_path}')
```
